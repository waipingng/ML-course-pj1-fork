```python
import pandas as pd
import re
```

```python
heart_df = pd.read_csv('../raw_data/heart_raw.csv')
```

```python
# Rename Data_Value column to Heart Disease Mortality
heart_df = heart_df.rename(columns={'Data_Value': 'Heart Disease Mortality', 'Stratification1': 'Sex', 'Stratification2':'ethnicity'})


# Display the first few rows to verify the change
heart_df.head()
```

```python

columns_drop = [
    'Data_Value_Footnote',
    'Data_Value_Type',
    'Class',
    'DataSource',
    'Data_Value_Footnote_Symbol',
    'StratificationCategory1',
    'Topic',
    'Data_Value_Unit',
    'StratificationCategory2',
]

heart_df = heart_df.drop(columns=columns_drop)

heart_df.head()
```

```python
heart_df.dropna()
```

```python
def restructure_csv(input_file):
    df = pd.read_csv(input_file)
    
    columns = df.columns.tolist()
    
    transformed_rows = []
    
    current_county = None
    current_state = None
    current_household_type = None
    
    for i, row in df.iterrows():
        first_column_value = str(row[columns[0]]).strip()
        
        county_match = re.match(r'(.+) County, (.+)', first_column_value)
        if county_match:
            current_county = county_match.group(1)
            current_state = county_match.group(2)
            continue
            
        if first_column_value in ['Households', 'Families', 'Married-couple families', 'Nonfamily households']:
            current_household_type = first_column_value
            continue
            
        if first_column_value == 'Estimate':
            new_row = {
                'County': current_county,
                'State': current_state,
                'Household_Type': current_household_type
            }
            
            for j in range(1, len(columns)):
                clean_col_name = columns[j].replace('Total!!', '').strip()
                new_row[clean_col_name] = row[columns[j]]
                
            transformed_rows.append(new_row)
    
    result_df = pd.DataFrame(transformed_rows)

    
    return result_df


```

```python
result_df = restructure_csv("../raw_data/income_2020.csv")  # For wide format
```

```python
result_df.head(10)
```

```python
filtered_df = result_df[result_df['Household_Type'] == 'Households']

filtered_df = filtered_df.reset_index(drop=True)

filtered_df.head()
```

```python
filtered_df.shape
```

```python
def clean_county(heart_df):
    df_cleaned = heart_df.copy()
    
    def remove_county(location):
        if isinstance(location, str):
            return location.replace(' County', '').replace('County ', '').replace(' County ', ' ')
        return location 
    
    df_cleaned['LocationDesc'] = df_cleaned['LocationDesc'].apply(remove_county)
    
    return df_cleaned

```

```python
heart_df_cleaned = clean_county(heart_df)

heart_df_cleaned.head(10)
```

```python
filtered_df_unique = filtered_df.drop_duplicates(subset=['County'])
merged_df = pd.merge(
    filtered_df_unique,
    heart_df_cleaned,
    left_on='County',
    right_on='LocationDesc',
    how='right' 
)

merged_df.head()
```
